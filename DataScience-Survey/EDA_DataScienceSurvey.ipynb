### Merging columns of Multiple Select Question & Creating new columns.

```python
df['Programming_Language'] = df[df.columns[4:19]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['IDE'] = df[df.columns[19:33]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Hosted_Notebook'] = df[df.columns[33:49]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Visualization_Libraries'] = df[df.columns[49:64]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['ML_Frameworks'] = df[df.columns[64:79]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Pre-Trained_Model_Weights'] = df[df.columns[79:89]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Best_Exp_Cloud_platForm'] = df[df.columns[96:108]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Cloud_Computing_Products'] = df[df.columns[109:114]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Data_Storage_Products'] = df[df.columns[114:122]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['DBMS'] = df[df.columns[122:138]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['BI_Tools'] = df[df.columns[138:153]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Managed_ML_Products'] = df[df.columns[153:166]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Auto_ML_Tools'] = df[df.columns[166:174]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Serve_ML_Models'] = df[df.columns[174:186]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['ML_Monitering_Tools'] = df[df.columns[186:201]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Ethical_AI_Products'] = df[df.columns[201:210]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
df['Hardware'] = df[df.columns[210:219]].apply(lambda x: ','.join(x.dropna().str.strip()),axis=1)
```

### Renaming values with large names replacing null/NaN with None.

```python
df.Country = df.Country.replace(
   ['United States of America','United Kingdom of Great Britain and Northern Ireland','Viet Nam','Iran, Islamic Republic of...', 'United Arab Emirates', 'Czech Republic', 'Hong Kong (S.A.R.)', 'I do not wish to disclose my location'],
   ['USA', 'UK', 'Vietnam', 'Iran',  'UAE', 'Czechia', 'Hong Kong', 'Not wish to disclose'])


df.Role = df.Role.replace(
   ['Data Analyst (Business, Marketing, Financial, Quantitative, etc)', 'Manager (Program, Project, Operations, Executive-level, etc)'],
   ['Data Analyst', 'Manager'])

df.Cloud_Expenses = df.Cloud_Expenses.replace(['$0 ($USD)', '$100,000 or more ($USD)'],
                          ['$0', '$100,000 or more'])

df.Best_Cloud_Experience = df.Best_Cloud_Experience.replace(['They all had a similarly enjoyable developer experience'],
                          ['Similar experience'])

df.IDE = df.IDE.replace(['', 'Visual Studio Code (VSCode)'],
               ['None', 'VSCode'])

df.fillna('None', inplace=True)

df = df.replace([''], 'None')
```
